```python
#Importing the necessary libraries
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

```python
#Search for the target protain (Coronavirus)
target = new_client.target
target_query = target.search('coronavirus')
targets = pd.DataFrame.from_dict(target_query)
display (targets)
                            
```

```python
#Assign the fourth entry to the selected target variable
selected_target = targets.target_chembl_id[6]
display(selected_target)
```

```python
'''Retrieving only the bioactivity data for coronavirus 3C-like proteinase(CHEMBL3927) that are reported as
IC50 values in nM(nanomolar)unit'''

activity = new_client.activity
res = activity.filter(target_chembl_id = selected_target).filter(standard_type = "IC50")
```

```python
df = pd.DataFrame.from_dict(res)
```

```python
df.standard_type.unique()
```

```python
#Converting the resulting bioactivity data into a CSV file
df.to_csv('bioactivity_data.csv', index=False) 
```

```python
df2 = pd.read_csv("bioactivity_data.csv")
```

```python
df2.sample()
```

```python
# Handling missing data
#Drop any compound having missing value for standard_value column

df2 = df[df.standard_value.notna()]
display(df2)
```

```python
df2.isnull().sum()
```

```python
#Labeling compounds as either being acvite, inactive or intermediate.
bioactivity_class = []
for i in df2.standard_value:
    if float(i) >=10000:
        bioactivity_class.append("inactive")
    elif float(i) <= 1000:
        bioactivity_class.append("active")
    else:
         bioactivity_class.append("intermediate")
#For a better runtime for the application we will use the alternative code below.
'''bioactivity_class = []
for i in df2.standard_value:
    value = float(i)
    bioactivity_class.append({
        True: 'inactive',
        False: 'intermediate'
    }.get(value <= 1000, 'active'))'''
```

```python
#Iterate the molecule_chembl_id to a list

mol_cid = []
for i in df2.molecule_chembl_id:
    mol_cid.append(i)
```

```python
#Iterate the canonical_smiles to a list

canonical_smiles = []
for i in df2.canonical_smiles:
    canonical_smiles.append(i)
```

```python

#Iterate the standard value to a list

standard_value = []
for i in df2.standard_value:
    standard_value.append(i)
```

```python
#Combining the four lists into a dataframe

data_tuples = list(zip(mol_cid, canonical_smiles, bioactivity_class, standard_value))
df3 = pd.DataFrame(data_tuples, columns=['molecule_chembl_id', 'canonical_smiles', 'bioactivity_class', 'standard_value'])
```
